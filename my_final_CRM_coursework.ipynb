```python
import yfinance as yf
!pip install yahoofinancials
from yahoofinancials import YahooFinancials
from datetime import datetime

# start and end dates
start = '2022-01-01'
end = '2025-01-01'

# Remove symbols from the list
excluded_symbols = ['BF.B', 'SOLS', 'Q', 'BRK.B']
symbols = [sym for sym in sp500_table['Symbol'].values.tolist() if sym not in excluded_symbols]

# Download stock price data
Data_stock_price = yf.download(symbols, start=start, end=end)
```

```python
Data_stock_price
```

```python
#calculate the daily returns and annual volatility of the stock
import math
daily_returns = Data_stock_price['Close'].pct_change()
annual_volatility = daily_returns.std() * math.sqrt(252)
```

```python
import yfinance as yf
import pandas as pd

# Download S&P 500 index (^GSPC) adjusted close prices
sp500_index_full_df = yf.download('^GSPC', start='2022-01-01', end='2025-01-01', auto_adjust=False)
sp500_index = sp500_index_full_df['Adj Close'].squeeze()

# Calculate daily returns
sp500_returns = sp500_index.pct_change()

# Calculate Beta
beta_values = {}
for ticker in daily_returns.columns:
    stock_returns = daily_returns[ticker]

    valid_idx = stock_returns.dropna().index.intersection(sp500_returns.dropna().index)

    if len(valid_idx) > 0 and stock_returns[valid_idx].std() > 0 and sp500_returns[valid_idx].std() > 0:
        corr = stock_returns[valid_idx].corr(sp500_returns[valid_idx])
        beta = corr * (stock_returns[valid_idx].std() / sp500_returns[valid_idx].std())
    else:
        beta = float('nan')
    beta_values[ticker] = beta

beta_series = pd.Series(beta_values)

#show values
print("Beta values:")
print(beta_series)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import pandas as pd

# beta_series
betas = pd.DataFrame(beta_series, columns=['Beta']).dropna()
betas['Stock_Name'] = betas.index


X = betas['Beta'].values.reshape(-1, 1)

# number of clusters
for n_clusters in range(2, 6):
    model = AgglomerativeClustering(n_clusters=n_clusters)
    labels = model.fit_predict(X)

    # labels to DataFrame
    betas['Cluster'] = labels

    # Scatter plot

    plt.scatter(betas['Beta'], np.arange(len(betas)), c=labels, cmap='rainbow', s=10)
    plt.xlabel('Beta')
    plt.ylabel('Beta')
    plt.title(f'{n_clusters} Clusters')
    plt.yticks([])
    plt.show()

    # show stocks
    cluster_counts = np.bincount(labels)
    for i in range(n_clusters):
        print(f"Cluster {i+1} has {cluster_counts[i]} stocks")
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Beta values
X = beta_series.dropna().values.reshape(-1, 1)

# Ward linkage
Z = linkage(X, method='ward')


# Plot Dendrogram
plt.figure(figsize=(15, 7))

dendrogram(
    Z,
    truncate_mode='lastp',
    p=30,
    leaf_rotation=90,
    leaf_font_size=9,
    show_contracted=True
)

plt.title("Dendrogram")
plt.xlabel("Stocks")
plt.ylabel("Euclidean Distance")

# lines
plt.axhline(y=3, color='orange', linestyle='--', label="3-Cluster Cut")
plt.axhline(y=4, color='purple', linestyle='--', label="4-Cluster Cut")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import pandas as pd

# Beta values
X = beta_series.dropna().values.reshape(-1, 1)

# clusters from 2 to 9
range_n_clusters = range(2, 9)
scores = []


for k in range_n_clusters:
    model = AgglomerativeClustering(n_clusters=k)
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)


# Plot Silhouette Score
plt.plot(list(range_n_clusters), scores, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# stocks with Beta and annual volatility values
common = beta_series.index.intersection(annual_volatility.index)

df = pd.DataFrame({
    'Stock': common,
    'Beta': beta_series.loc[common].values,
    'AnnualVolatility': annual_volatility.loc[common].values
}).dropna()

# K-Means number
K = 3
df['Cluster'] = KMeans(n_clusters=K, random_state=42, n_init=10)\
                .fit_predict(df[['Beta', 'AnnualVolatility']])

# Scatter plot
plt.figure(figsize=(21, 9))
plt.scatter(df['Beta'], df['AnnualVolatility'], c=df['Cluster'], cmap='viridis', s=40)

plt.xlabel("Beta")
plt.ylabel("Annual Volatility")
plt.show()
print(df.sort_values("Cluster"))


```
